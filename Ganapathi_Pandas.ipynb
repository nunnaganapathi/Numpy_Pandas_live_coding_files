```python
#Creating the dataset
data1 ={
    'Name':['Jai','Princi','Gaurav','Anuj'],
    'Age':[27,24,22,32],
    'Address':['Delhi','Kanpur','Allahabad','Kannauj'],
    'Qualification':['Msc','MA','MCA','MBA'] }

data2 ={
    'Name':['Abhi','Ayushi','Dhiraj','Hitesh'],
    'Age':[17,14,12,52],
    'Address':['Nagpur','Kanpur','Allahabad','Kannauj'],
    'Qualification':['Btech','B.A','Bcom','B.hons']
}

#Converting to dataframes

df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[4,5,6,7])

print(df,'\n\n',df1)
```

```python
#concatinating the two dataframes

res=pd.concat([df,df1])
print(res)
```

```python
#Column Conactination
res=pd.concat([df,df1],axis=1)
print(res)
```

```python
'''
By default -> we have 0 and 1 values for axis
if we mention axis=0 , the operation would be performed by row wise. By default the value of axis is 0
if we mention axis=1 , the operation would be performed by column wise

'''
```

```python
#Column Concotination.
res=pd.concat([df,df1],axis=1, join='inner')

print(res)
```

```python
#Column Concotination.
res=pd.concat([df,df1],axis=1, join='outer')

print(res)
```
