```python
#Creating a dataframe
df1=pd.DataFrame({
    'Key':['K0','K1','K2','K3'],
    'Key1':['K0','K1','K2','K3'],
    'Name':['Jai','Princi','Gaurav','Anuj'],
    'Age':[23,24,25,26]

})
df2=pd.DataFrame({
    'Key':['K0','K1','K2','K3'],
    'Key1':['K0','K1','K2','K3'],
    'Address':['Nagpur','Kanpur','Allahabad','Aurangabad'],
    'Qualification':['Btech','B.A','Bcom','B.Hons']

})

print(df1,'\n\n',df2)
```

```python
#Merge with multiple keys (compsite keys)
res=pd.merge(df1,df2,on=['Key','Key1'])
print(res)
```

```python
#Creating a dataframe
df1=pd.DataFrame({
    'Key':['K0','K1','K2','K3'],
    'Key1':['K0','K1','K0','K1'],
    'Name':['Jai','Princi','Gaurav','Anuj'],
    'Age':[23,24,25,26]

})
df2=pd.DataFrame({
    'Key':['K0','K1','K2','K3'],
    'Key1':['K0','K0','K0','K0'],
    'Address':['Nagpur','Kanpur','Allahabad','Aurangabad'],
    'Qualification':['Btech','B.A','Bcom','B.Hons']

})

print(df1,'\n\n',df2)
```

```python
#(K0,K0) and (K2,K0) are the common (key,key1) Combinations
```

```python
#Merge with multiple keys (Composite key)
res=pd.merge(df1,df2,on=['Key','Key1'])
print(res)
```

```python
#Merge with multiple keys (Composite key)
res=pd.merge(df1,df2,how='left', on=['Key','Key1'])
print(res)
```

```python
#Merge with multiple keys (Composite key)
res=pd.merge(df1,df2,how='right', on=['Key','Key1'])
print(res)
```

```python
#Merge with multiple keys (Composite key)
res=pd.merge(df1,df2,how='outer', on=['Key','Key1'])
print(res)
```

```python
#Merge with multiple keys (Composite key)
res=pd.merge(df1,df2,how='inner', on=['Key','Key1'])
print(res)
```

```python
#Pivot tables

#Creating the dataframe
df=pd.DataFrame({
    'Product': ['Carrot','Bracoli','Banana','Banana',
                'Beans','Orange','Bracolo','Banana'],
    'Category':['Vegetable','Vegetable','Fruit','Fruit',
    'Vegetable','Fruit','Vegetable','Fruit'],
    'Quantity': [8,5,3,4,5,9,11,8],
    'Amount'  : [270,240,620,384,626,610,620,190]
})
df.head()
```

```python
#Pivot tables with sum Aggregation
pivot=df.pivot_table(
    index='Product',
    values='Amount',aggfunc='sum')
print(pivot)
```

```python
#Pivot tables with sum Aggregation
pivot=df.pivot_table(
    index='Category',
    values='Amount',aggfunc='sum')
print(pivot)
```

```python
#Pivot tables with sum Aggregation
pivot=df.pivot_table(
    index=['Product','Category'],
    values='Amount',
    aggfunc='sum')
print(pivot)
```

```python
#Pivot tables with sum Aggregation
pivot=df.pivot_table(
    index=['Product','Category'],
    values=['Amount','Quantity'],
    aggfunc='sum')
print(pivot)
```

```python
#Pivot tables with sum Aggregation
pivot=df.pivot_table(
    index=['Product','Category'],
    values=['Amount','Quantity'],
    aggfunc=['sum','median'])
print(pivot)
```
